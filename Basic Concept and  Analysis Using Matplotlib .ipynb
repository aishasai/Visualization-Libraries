Matplotlib is an  visualization library in Python for 2D and 3D graphics.
Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.
It was introduced by John Hunter in the year 2002. 


```python
import matplotlib.pyplot as plt #importing library as plt
import numpy as np
```

```python
#Lets go through some basic plots with the data set TIPS
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x,y);
```

Adding comments to the graph

```python
plt.plot(x, y, 'b--')
# b indicate what color should be the plot(r,g etc) and --(*,o etc) the model
plt.xlabel('Title for axis x')
plt.ylabel('Title for axis y')
plt.title('Graph Title')
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'b*-') 
plt.subplot(1,2,2)
plt.plot(y, x, 'g*');
```

```python
plt.figure(figsize=(8,4))# figsize is used to adjust the plotted figure size
plt.plot(x,y)
```

How we can add more than one axis in a graph compare.

```python

fig = plt.figure()# Creates blank canvas

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inner axes

axes1.plot(x, y, 'r*')
axes1.set_xlabel('axesx1')
axes1.set_ylabel('axesy1')
axes1.set_title('Plot1')

# Insert Figure Axes 2
axes2.plot(y, x, 'r--')
axes2.set_xlabel('axesx2')
axes2.set_ylabel('axesy2')
axes2.set_title('plot2');
```

```python
#Syntax to add Label and Legend
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```
